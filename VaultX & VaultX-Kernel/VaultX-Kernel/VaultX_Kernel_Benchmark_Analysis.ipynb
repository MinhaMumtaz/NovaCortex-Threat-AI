# VaultX-Kernel Benchmark Analysis

This notebook documents the benchmark methodology used to evaluate
the performance of VaultX-Kernel during file encryption and decryption.

## Benchmark Methodology

- Test File Size: 1 MB
- Encryption Mode: AES-256-GCM
- Key Derivation: PBKDF2 (100,000 iterations)
- Timing Source: CLOCK_MONOTONIC
- Environment: Linux (VMware)


The benchmark is implemented in C using high-resolution timers:

- clock_gettime(CLOCK_MONOTONIC)
- External execution of vaultx_kernel
- Password piped via stdin


## Benchmark Results (Observed)

Encryption Time: ~0.03–0.04 seconds
Decryption Time: ~0.03–0.04 seconds

These results demonstrate:
- Minimal overhead
- Efficient EVP usage
- Low-latency cryptographic operations


## Interpretation

VaultX-Kernel significantly outperforms high-level implementations
due to:

- Direct OpenSSL EVP usage
- Absence of interpreter overhead
- Efficient memory handling

This makes it suitable for system-level security tooling.
